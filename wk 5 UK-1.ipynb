```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns


### Part1

df = pd.read_csv("C:\\Users\\dell\\Desktop\\amz_uk_price_prediction_dataset.csv", encoding='latin-1')
category_frequency = df['category'].value_counts()
print("f top 5 most listed product categories are:")
print (category_frequency.head(5))

plt.figure(figsize=(10,6))
category_frequency.head(10).plot(kind='bar')
plt.title("Different categories product distribution")
plt.xlabel("product category")
plt.ylabel("frequency")
plt.xticks(rotation =45)
plt.show()

plt.figure(figsize=(8,8))
category_frequency.head(5).plot(kind='pie',autopct='%1.1f%%')
plt.title("f top 5 most listed product categories")
plt.ylabel("")
plt.show()

### yes from the pie chart we find Sports & Outdoor dominate the listing


### Part2

df['price'] = pd.to_numeric(df['price'],errors ='coerce')
pd.options.display.float_format='{:.2f}'.format
df= df.dropna(subset=['price'])

print(df['price'].mean())
print(df['price'].median())
mode_price = stats.mode(df['price'])[0]
print(f" model price is: {mode_price}")

### 2
variance_price = df['price'].var()
std_dev_price = df['price'].std()
price_range = df['price'].max() -df['price'].min()
interquartile_range = np.percentile(df['price'],75) - np.percentile(df['price'],25)
print(f"variance_price is {variance_price},std_dev_price is {std_dev_price},price_range is {price_range},interquartile_range is {interquartile_range}")

# ### 3
plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins=30,kde=True)
plt.title('Price distribution')
plt.xlabel('price')
plt.ylabel('frequency')
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x=df['price'])
plt.title('price boxplot')
plt.xlabel('price')
plt.show()

### Part 3

## 1
df.loc[(df['stars'] < 0) | (df['stars'] > 5),'stars'] =0

print(df['stars'].mean())
print(df['stars'].median())
print(df['stars'].mode())
# till now i didn't find any trend when customer rate products

## 2
variance_stars = df['stars'].var()
std_dev_stars = df['stars'].std()
stars_range = df['stars'].max() -df['stars'].min()
interquartile_range = np.percentile(df['stars'],75) - np.percentile(df['stars'],25)
print(f"variance_stars is {variance_stars},std_dev_stars is {std_dev_stars},stars_range is {stars_range},interquartile_range is {interquartile_range}")
# we will reply this question basing on following reply


## 3
skewness_rating = df['stars'].skew()
kurtosis_rating = df['stars'].kurt()
print(f"skewness_rating for rating: {skewness_rating} and kurtosis_rating for rating is {kurtosis_rating }")
#

## 4
plt.figure(figsize=(10,6))
sns.histplot(df['stars'], bins=20,kde =True)
plt.title("Distribution for product ratings")
plt.xlabel("Rating")
plt.show()
# yes from the histogram we found more than half customers didn't offer any ratings
```
